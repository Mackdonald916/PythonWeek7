#### Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Load the dataset

```python
df = pd.read_csv("student_depression_dataset.csv")
```

#### Display the first few rows of the dataset

```python
print(df.head())
```

#### Explore the dataset structure

```python
print("\nDataset Info:")
print(df.info())
```

#### Check for missing values

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
df = df.dropna()
print("\nCleaned Dataset Info:")
print(df.info())
```

#### Grouping by a categorical column and computing the mean of 'Age'

```python
grouped_data = df.groupby('Gender')['Age'].mean()
print("\nAverage Age by Gender:")
print(grouped_data)
```

#### Identifying patterns (e.g., correlation between CGPA and Depression)

```python
print("\nCorrelation between CGPA and Depression:")
print(df[['CGPA', 'Depression']].corr())
```

#### Line chart: Trends over time

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Age'], df['CGPA'], marker='o', linestyle='-')
plt.title('CGPA vs Age', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('CGPA', fontsize=12)
plt.grid(True)
plt.show()
```

#### Bar chart: Average CGPA per gender

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Gender', y='CGPA', data=df)
plt.title('Average CGPA by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Average CGPA', fontsize=12)
plt.show()
```

#### Histogram: Distribution of CGPA

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['CGPA'], bins=10, kde=True)
plt.title('Distribution of CGPA', fontsize=16)
plt.xlabel('CGPA', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

#### Scatter plot: CGPA vs Age

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='CGPA', data=df)
plt.title('CGPA vs Age', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('CGPA', fontsize=12)
plt.show()
```
